# `pyroomacoustics` demo for source directivities

In this IPython notebook, we demonstrate how to add a source with directivities using pyroomacoustics and look at how this affects the RIR plots and output signals:

```python
import pyroomacoustics as pra
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
import IPython
from pyroomacoustics.directivities import DirectivityPattern, DirectionVector, CardioidFamily
```

```python
def make_room(dimensions, material, fs, max_order):
    return pra.ShoeBox(
            p = dimensions,
            materials=pra.Material(material),
            fs=fs,
            max_order=max_order,
        )
```

## Original signal

```python
fs, signal = wavfile.read("arctic_a0001.wav")
print("Original WAV:")
IPython.display.Audio(signal, rate=fs)
```

## CASE 1: Omnidirectional source

```python
# make room by specifying the parameters
room1 = make_room(dimensions=[6.5, 4, 4], material=0.07, fs=16000, max_order=17)

# add source signal with omnidirectional directivity
PATTERN = DirectivityPattern.OMNI
ORIENTATION = DirectionVector(azimuth=0, colatitude=0, degrees=True)
directivity_1 = CardioidFamily(orientation=ORIENTATION, pattern_enum=PATTERN)

room1.add_source([1,2,2], signal=signal, directivity=directivity_1)
fig1, ax1 = room1.plot()
ax1.set_xlim([0, 6.5])
ax1.set_ylim([0, 4])
ax1.set_zlim([0, 4])

# add microphone  
room1.add_microphone([3.5,2,2])

# 3-D plot
azimuth = np.linspace(start=0, stop=360, num=361, endpoint=True)
colatitude = np.linspace(start=0, stop=180, num=180, endpoint=True)

print("3-D plot for case 1:")
directivity1.plot_response(azimuth=azimuth, colatitude=colatitude, degrees=True, ax=ax1, offset=[1,2,2])
plt.show()

# 2-D plot
corners = np.array([[0,4], [6.5,4], [6.5,0], [0,0]]).T 
room = pra.Room.from_corners(corners)
room.add_microphone([3.5,2])
fig, ax = room.plot()

print("2-D plot for case 1:")
directivity1.plot_response(azimuth=azimuth, colatitude=None, degrees=True, ax=ax, offset=[1,2])
plt.show()

# plot RIR
print("RIR plot for case 1:")
room1.plot_rir()
plt.show()

# display output signal for case 1
room1.simulate()
print("Simulated propagation to mic for case 1:")
IPython.display.display(IPython.display.Audio(room1.mic_array.signals[0,:], rate=fs))
```

## CASE 2: Source with directivity pointing away from the source

```python
# make room with the same parameters as before
room2 = make_room(dimensions=[6.5, 4, 4], material=0.07, fs=16000, max_order=17)

# add source signal to the same point as before with directivity pointing away from the microphone
PATTERN = DirectivityPattern.FIGURE_EIGHT
ORIENTATION = DirectionVector(azimuth=90, colatitude=90, degrees=True)
directivity_2 = CardioidFamily(orientation=ORIENTATION, pattern_enum=PATTERN)

room2.add_source([1,2,2], signal=signal, directivity=directivity_2)
fig2, ax2 = room2.plot()
ax2.set_xlim([0, 6.5])
ax2.set_ylim([0, 4])
ax2.set_zlim([0, 4])

# add microphone to the same point as before
room2.add_microphone([3.5,2,2])

# 3-D plot
azimuth = np.linspace(start=0, stop=360, num=361, endpoint=True)
colatitude = np.linspace(start=0, stop=180, num=180, endpoint=True)

print("3-D plot for case 2:")
directivity2.plot_response(azimuth=azimuth, colatitude=colatitude, degrees=True, ax=ax2, offset=[1,2,2])
plt.show()

# 2-D plot
corners = np.array([[0,4], [6.5,4], [6.5,0], [0,0]]).T 
room = pra.Room.from_corners(corners)
room.add_microphone([3.5,2])
fig, ax = room.plot()

print("2-D plot for case 2:")
directivity2.plot_response(azimuth=azimuth, colatitude=None, degrees=True, ax=ax, offset=[1,2])
plt.show()

# plot RIR
print("RIR plot for case 2:")
room2.plot_rir()
plt.show()

# display output signal for case 2
room2.simulate()
print("Simulated propagation to mic for case 2:")
IPython.display.Audio(room2.mic_array.signals[0,:], rate=fs)
```

## CASE 3: Source with directivity pointing towards from the source
